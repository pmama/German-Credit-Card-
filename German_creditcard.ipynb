### **Checking and Removing Duplicates**

```python
# Remove duplicates

def drop_dup(df):
    if df.duplicated().any() == True:
        print('The total duplicate row before removing duplicate:', df.duplicated().sum())
        df.drop_duplicates(inplace=True , keep = 'last')  # Remove duplicates
        df = df.reset_index(drop=True) #Reset the index
        print('The total duplicate row after removing duplicate:', df.duplicated().sum(), ' \nshape of dataset after removing duplicate columns :',df.shape)
    else:
        return 'No duplicate entries'
drop_dup(df)
```

# Outlier Treatment : 

```python
def outlier(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3-Q1
    upper_bound = Q3 + 1.5*IQR
    lower_bound = Q1 - 1.5*IQR
    return data.clip(upper_bound, lower_bound)

df['amount'] = outlier(df['amount'])
num_dist('amount',target ='default')
```

## Splitting into target and independent features

```python
target = "default"
x = df.drop(columns=target)
y = df[target]
```

## Feature Scaling

```python
#StandardScaler in dataframe
sc = StandardScaler()
sc_x = pd.DataFrame(sc.fit_transform(x) , columns=x.columns)
sc_x.head()
```

## Imbalance treatment

```python
from imblearn.over_sampling import RandomOverSampler
over = RandomOverSampler()
x_over, y_over = over.fit_resample(sc_x,y)
```

```python
y_over.value_counts().plot(kind='bar', color=['skyblue', 'lightgreen'])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_over,y_over,test_size=0.25,random_state=123,  stratify = y_over)
```

## Model comparision

```python
m1 = LogisticRegression()
m2 = DecisionTreeClassifier(max_depth =11, criterion='log_loss' )
m3 = BaggingClassifier(estimator=RandomForestClassifier())
m4 = RandomForestClassifier(max_depth = 10, criterion = 'log_loss', random_state =123 )
m5 = SVC(kernel ='rbf', C =.3,)


models = [('LogisticRegression', m1), 
           ('DecisionTreeClassifier', m2), 
           ('BaggingClassifier', m3), 
          ('RandomForestClassifier',m4), 
           ('SupportVectorClassifier',m5) 
        ]


from tabulate import tabulate
print(tabulate(models, headers=['Model name', 'Sklearn models']),)
```
